```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

SOAL A

```python
df = pd.read_csv('stock.csv', sep=';')
df.head(13)
```

```python
# Menampilkan histogram untuk atribut "Purchase Price" menggunakan Seaborn
plt.figure(figsize=(20, 6))
plt.hist(df['Purchase Price'], bins=55, edgecolor='brown')
plt.xlabel('Purchase Price')
plt.ylabel('Count')
plt.title('Histogram of Purchase Price')
plt.show()
```

```python
# Menampilkan histogram untuk atribut "Current Price" menggunakan Seaborn
plt.figure(figsize=(10, 8))
plt.hist(df['Current Price'], bins=55, edgecolor='brown')
plt.xlabel('Current Price')
plt.ylabel('Count')
plt.title('Histogram of Current Price')
plt.show()
```

```python
# Menampilkan bar chart untuk atribut "Purchase Price" menggunakan Seaborn
plt.figure(figsize=(22, 6))
plt.hist(df['Purchase Price'], bins=55, edgecolor='brown')
plt.xlabel('Purchase Price')
plt.ylabel('Count')
plt.title('Bar Chart of Purchase Price')
plt.show()
```

```python
# Membaca file CSV
df = pd.read_csv('stock.csv', sep=';')

# Mengurutkan data berdasarkan Dividend/Share secara menurun
sorted_stocks = df.sort_values(by='Dividend/Share', ascending=False) .head(5)[[
  'Stock Name','Symbol'
]]

# Menampilkan atribut Stock Name dan Symbol dari lima saham teratas

print(sorted_stocks)
```
